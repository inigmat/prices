## Чтение файлов из папки поочередно


```python
import pandas as pd
import os

repo_folder = '/content/prices/files/raw'

dataframes_dict = {}

# Прочитайте файлы из папки поочередно
for filename in os.listdir(repo_folder):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(repo_folder, filename)
        df = pd.read_excel(file_path)
        # Сохраняем DataFrame в словарь с соответствующим названием файла
        dataframes_dict[filename] = df
```

## Наименования таблица

```python
df_names = list(dataframes_dict.keys())
df_names
```

## Поиск по слову

```python
# Искомое слово
search_query = "KAO"

# Поиск по всем таблицам в словаре dataframes_dict
search_results = {}
for table_name, df in dataframes_dict.items():
    # Применяем метод str.contains для каждой ячейки DataFrame
    search_result = df[df.apply(lambda row: row.astype(str).str.contains(search_query, case=False)).any(axis=1)]
    if not search_result.empty:
        search_results[table_name] = search_result
```

## Вывод результатов одним списком

```python
# Вывод результатов
print("Результаты поиска:")
for table_name, df in search_results.items():
    print(f"Из таблицы '{table_name}':")
    for index, row in df.iterrows():
        print(row.dropna())
    print()
```
