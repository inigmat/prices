```python
import pandas as pd
kosmart_mp = pd.read_excel('kosmart_07_mp.xlsx')
# Сохраняем DataFrame в словарь с соответствующим названием файла
kosmart_mp.head(5)
```

```python
headers = kosmart_mp.iloc[2]

kosmart_mp  = pd.DataFrame(kosmart_mp.values[4:], columns=headers)

# Выбранные столбцы, которые нужно преобразовать
selected_columns = ['Цена оптовая', 'Цена РРЦ']

# Удаление строк с пустыми значениями цен
kosmart_mp = kosmart_mp.dropna(subset=selected_columns)

# Удаление пробелов и преобразование в числа формата int
for col in selected_columns:
    #kosmart_mp.loc[:, col] = kosmart_mp.loc[:, col].str.replace(' ', '').astype(int)
    kosmart_mp[col] = kosmart_mp[col].str.replace(' ', '').astype(int)
```

```python
kosmart_mp.head(5)
```

# Объединение

```python
import pandas as pd

# Применение аналога VLOOKUP
kosmart_merged = kosmart_mp.merge(raiting_mp, on='Номенклатура', how='left')
```

```python
# Выбранные столбцы для удаления
selected_columns = ['Описание', 'Заказ, шт', 'Сумма, руб.', 'Итого', 'Фото', ' ']
kosmart = kosmart_merged.drop(columns=selected_columns)

# Удаление строк с пустыми значениями цен
kosmart = kosmart.dropna(subset=['Кол-во проданных позиций'])
```

```python
# Фукнция для расчета скидки
def subtract_percent(number):
    percent = 15
    deduction = number * percent / 100
    result = number - deduction
    return result
```

```python
#kosmart_c['Цена РРЦ'].astype('int')
#kosmart_c['Цена оптовая'].astype('int')
kosmart['Маржин'] = kosmart['Цена РРЦ']/kosmart['Цена оптовая']
kosmart['Оптовая со скидкой'] = subtract_percent(kosmart['Цена оптовая'])
```

# С озон

## Код для ввода данных

```python
%%time
import pandas as pd
import ipywidgets as widgets
from IPython.display import clear_output, display

columns = {
    'Name': 'Номенклатура',
    'Seller Price': 'Цена продавца',
    'Kosmart Price': 'Цена KM со скидкой',
    'Calc Price': 'Цена в калькуляторе',
    'FBO Cost': 'FBO Общие затраты на Ozon, за шт.',
    'FBS Cost': 'FBS Общие затраты на Ozon, за шт.',
    'URL': 'URL'
}

# Создание пустого DataFrame с заданными колонками
ozon = pd.DataFrame(columns=columns.keys())

# Список для хранения виджетов
input_widgets = []

# Создание виджетов для ввода данных
for col_name, col_desc in columns.items():
    if col_name == 'Name' or col_name == 'URL':
        widget = widgets.Text(description=col_desc)
    else:
        widget = widgets.IntText(description=col_desc)
    input_widgets.append(widget)

# Кнопка для добавления строки
add_button = widgets.Button(description='Add Row')
output_area = widgets.Output()

# Обработчик нажатия кнопки
def add_row(_):
    global ozon
    new_row = {col_name: widget.value for col_name, widget in zip(columns.keys(), input_widgets)}
    ozon = pd.concat([pd.DataFrame(new_row, index=[0]), ozon.loc[:]], ignore_index=True)

    # Очищаем и выводим только последнюю добавленную строку
    with output_area:
        clear_output(wait=True)
        display(ozon.head(1))

    for widget in input_widgets:
        if isinstance(widget, widgets.IntText):
            widget.value = 0
        else:
            widget.value = ''

# Привязка обработчика к кнопке
add_button.on_click(add_row)
```

## Ввод данных

```python
# Отображение виджетов, кнопки и области вывода
display(*input_widgets, add_button, output_area)
```
